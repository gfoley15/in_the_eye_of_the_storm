```python
import pandas as pd
import polars as pl
import os
```

```python
# Read tracker CSV in for data filtering and cleansing
directory = r'C:\Users\gfoley\Downloads'
filename = 'ibtracs.NA.list.v04r01.csv'

# Create the full file path
file_path = os.path.join(directory, filename)

# Load the CSV file into a Polars DataFrame
tracker_df = pl.read_csv(file_path)
tracker_df.head()
```

```python
# Convert Polars DataFrame to pandas DataFrame
tracker_df = tracker_df.to_pandas()
```

```python
# Create list of most intense hurricanes provided by clean_hurricane_data
major_canes = [
    'EASY', 'KING', 'CAROL', 'HAZEL', 'AUDREY', 'HELENE', 'GRACIE',
    'DONNA', 'CARLA', 'BETSY', 'BETSY', 'ALMA', 'BEULAH', 'CAMILLE',
    'CELIA', 'ALICIA', 'ELENA', 'HUGO', 'ANDREW', 'ANDREW', 'EMILY',
    'OPAL', 'FRAN', 'BRET', 'CHARLEY', 'IVAN', 'JEANNE', 'DENNIS',
    'KATRINA', 'RITA', 'WILMA', 'HARVEY', 'IRMA', 'MICHAEL', 'LAURA',
    'ZETA', 'IDA', 'IAN', 'IDALIA'
]
len(major_canes)

# Remove rows where the 'NAME' column does not match any of the specified major canes
tracker_df = tracker_df[tracker_df['NAME'].str.upper().isin(major_canes)]
```
